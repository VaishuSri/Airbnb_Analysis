```python
#Extracting data
Extracted_Data = []
for i in col.find():
    data = {
    "Id": i['_id'],
    "listing_url": i['listing_url'],
    "name": i['name'],
    "description": i['description'],
    "house_rules": i.get('house_rules'),  
    "property_type": i['property_type'],
    "room_type": i['room_type'],
    "bed_type": i['bed_type'],
    "minimum_nights": int(i['minimum_nights']),
    "maximum_nights": int(i['maximum_nights']),
    "cancellation_policy": i['cancellation_policy'],
    "accommodates": i['accommodates'],
    "bedrooms": i.get('bedrooms'),  
    "beds": i.get('beds'),
    "availability": i['availability']['availability_365'],
    "price": i['price'],
    "security_deposit": i.get('security_deposit'),  
    "cleaning_fee": i.get('cleaning_fee'),  
    "extra_people": i['extra_people'],
    "guests_included": i['guests_included'],
    "number_of_reviews": i['number_of_reviews'],
    "review_scores": i.get('review_scores'),  
    "amenities": i['amenities'],
    "host_name": i['host']['host_name'],
    "host_id": i['host']['host_id'],
    "street": i['address']['street'],
    "country": i['address']['country'],
    "country_code": i['address']['country_code'],
    "location_type": i['address']['location']['type'],
    "longitude": i['address']['location']['coordinates'][0],
    "latitude": i['address']['location']['coordinates'][1],
    "is_location_exact": i['address']['location']['is_location_exact']
    }
    Extracted_Data.append(data)

df1=pd.DataFrame(Extracted_Data)

#Changing datatype
df1.price = df1.price.astype(str).astype(float)
df1.security_deposit = df1.security_deposit[~df1.security_deposit.isna()].astype(str).astype(float)
df1.cleaning_fee = df1.cleaning_fee[~df1.cleaning_fee.isna()].astype(str).astype(float)
df1.extra_people = df1.extra_people.astype(str).astype(float)
df1.guests_included = df1.guests_included.astype(str).astype(float)
#Changing dtype for review score
# Assuming your DataFrame is named df
column_name = 'review_scores'
# Check if the specified column contains dictionaries
if isinstance(df1[column_name].iloc[0], dict):
    # Convert each dictionary into a tuple
    df1[column_name] = df1[column_name].apply(tuple)
# Assuming your DataFrame is named df
# First, let's clean up the review_scores column
# For example, if it contains non-integer values or missing values, you might do something like this:

# Replace any non-integer values with NaN
df1['review_scores'] = pd.to_numeric(df1['review_scores'], errors='coerce')

# Handle missing values (NaN) according to your preference, for example, by filling them with a specific value like 0:
df1['review_scores'].fillna(0, inplace=True)

# Now, convert the column to integer type
df1['review_scores'] = df1['review_scores'].astype(int)

#host_id
df1.host_id=df1.host_id.astype(str).astype(int)

df1.info()

#Finding Missing value
df1.isna().sum()

#Filling missing values
df1['beds']=df1['beds'].fillna(df1['beds'].median)
df1['bedrooms']=df1['bedrooms'].fillna(df1['bedrooms'].mode)
df1['security_deposit']=df1['security_deposit'].fillna(df1['security_deposit'].median)
df1['cleaning_fee']=df1['cleaning_fee'].fillna(df1['cleaning_fee'].median)

#Finding missing text data
import pandas as pd

# Assuming df1 is your DataFrame
# Check for empty spaces in the DataFrame
empty_spaces = df1.isin(['']).sum()

# Display columns with empty spaces
print(empty_spaces)

# Filling Empty values in Description and House rules columns
df1['description'].replace(to_replace='', value='No Description Provided', inplace=True)
df1['house_rules'].replace(to_replace='', value='No House rules Provided', inplace=True)
df1['name'].replace(to_replace='', value='Not Available', inplace=True)

#Finding missing text data
import pandas as pd

# Assuming df1 is your DataFrame
# Check for empty spaces in the DataFrame
empty_spaces = df1.isin(['']).sum()

# Display columns with empty spaces
print(empty_spaces)
#Checking size of df before duplicating
print(df1.shape)
# Iterate through each column in the DataFrame
for column in df1.columns:
    # Check if any element in the column is of type list
    if any(isinstance(item, list) for item in df1[column]):
        print(f"Column '{column}' contains lists.")
        
        # Convert lists to strings
        df1[column] = df1[column].apply(lambda x: ','.join(x) if isinstance(x, list) else x)
        print("Lists converted to strings")

# Now you can safely drop duplicates
df1 = df1.drop_duplicates()
print(df1.shape)
```
